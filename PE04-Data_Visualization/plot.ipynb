```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PE04_winereview.csv')
"""
This function reads a CSV file with wine review data and shows how often each score appears using a bar graph.
Steps:
1. It loads the data from PE04_winereview.csv, counts how many times each score shows up and sorts the scores in order.
2. It makes a bar graph with orange bars showing the frequency of each score and adds a title and labels to make the graph easy to understand.
3. It shows the final graph.
This helps us see which wine scores are the most common.
"""
points_counts = df['points'].value_counts().sort_index() # count frequency of each unique point score
# Plot the bar graph
plt.figure(figsize=(10, 6))
points_counts.plot(kind='bar', color='orange')
plt.title('Wine Review Score')
plt.xlabel('Points')
plt.ylabel('Frequency')
plt.tight_layout()  
plt.grid(False)     
plt.show()
```

```python
df = pd.read_csv('PE04_winereview.csv')
"""
This function reads wine review data and creates two histograms to show how often different scores appear.
Steps:
1. Loads the data from PE04_winereview.csv, gets the 'points' column and removes any missing values and creates two side-by-side plots for comparison.
2. The first histogram groups scores using a bin size of 10 and the second histogram uses a bin size of 3 for more detailed groups (e.g., 8083, 8386, etc.).
3. Both plots use blue bars with black edges, and labels are added to make them easy to understand.
This helps compare how different bin sizes affect the view of score distribution.
"""
points = df['points'].dropna() # extract the points column for histogram plotting
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Plot 1: Bin size = 10
axes[0].hist(points, bins=range(80, 101, 10), color='dodgerblue', edgecolor='black')
axes[0].set_title('Wine score histogram\nBin size 10')
axes[0].set_xlabel('Points')
axes[0].set_ylabel('Frequency')

# Plot 2: Bin size = 3
axes[1].hist(points, bins=range(80, 101, 3), color='dodgerblue', edgecolor='black')
axes[1].set_title('Wine score histogram\nBin size 3')
axes[1].set_xlabel('Points')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('PE04_winereview.csv')
"""
This function reads wine review data and creates a histogram to show how often different wine prices appear.
Steps:
1. Loads the data from PE04_winereview.csv, cleans the price column by removing missing or non-numeric values and filters out prices to avoid extreme outliers.
2. Creates a histogram with a bin size of 5 and uses green bars with black edges for better visibility.
3. Adds a title and labels to explain the x-axis and y-axis.
This plot helps to understand the overall distribution of wine prices in the dataset in a clear and simplified way.
"""
prices = pd.to_numeric(df['price'], errors='coerce').dropna() # clean the price column
filtered_prices = prices[prices <= 200]  # Keep prices up to $200

# Plot histogram with bin size 5
plt.figure(figsize=(10, 6))
plt.hist(filtered_prices, bins=range(0, 205, 5), color='mediumseagreen', edgecolor='black')

# Labels and title
plt.title('Wine Price Distribution (Bin size 5)')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
